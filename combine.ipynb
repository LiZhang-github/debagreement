```python
data = pd.read_csv("/content/drive/MyDrive/code/bert_related/labeled_data.csv", index_col = False)
data = data[data['agreement_fraction']==1.000000]
bb = data[["author_parent","author_child",'label','datetime','subreddit']]

```

```python
# combine embedding from sentence transformer

sen_par = pd.read_csv('/content/drive/MyDrive/code/bert_related/new_sentence_parent.csv',index_col=[0])
sen_chi = pd.read_csv('/content/drive/MyDrive/code/bert_related/sentence_child.csv',index_col=[0])

sen_par['sen_par'] = np.array(sen_par.to_numpy()).tolist()
sen_chi['sen_chi'] = np.array(sen_chi.to_numpy()).tolist()
aa = pd.concat([sen_par['sen_par'], sen_chi['sen_chi']], axis=1)

aaa = aa.reset_index()    
bbb = bb.reset_index() 

cc = pd.concat([bbb, aaa], axis=1)
print(cc)


```

```python
# combine embedding from bert

bert_par = pd.read_csv('/content/drive/MyDrive/code/bert_related/bert_parent.csv',index_col=[0])
bert_chi = pd.read_csv('/content/drive/MyDrive/code/bert_related/bert_child.csv',index_col=[0])

bert_par['bert_par'] = np.array(bert_par.to_numpy()).tolist()
bert_chi['bert_chi'] = np.array(bert_chi.to_numpy()).tolist()
aa = pd.concat([bert_par['bert_par'], bert_chi['bert_chi']], axis=1)

aaa = aa.reset_index()    
bbb = bb.reset_index() 

cc = pd.concat([bbb, aaa], axis=1)
```

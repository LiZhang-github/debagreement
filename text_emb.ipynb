```python
# 加载roberta-large-nli-stsb-mean-tokens
# model = SentenceTransformer("araphrase-MiniLM-L6-v2")
model = SentenceTransformer("roberta-large-nli-stsb-mean-tokens")
sentences = ["I am Xiao Liu.", "I come from KEG, THU."] 
```

```python
sentence_embeddings = model.encode(sentences)
print(sentence_embeddings.shape)
```

```python
# 下载预训练模型，这可能需要较长的时间。中文可使用bert-base-chinese
tokenizer = AutoTokenizer.from_pretrained("bert-base-cased") 
model = AutoModel.from_pretrained("bert-base-cased")

tokenizer.encode("I am from KEG, Tsinghua University.", return_tensors="pt")
# A sequence with torch.Size([1, 13]) : 
# [[‘[CLS]', 'I', 'am', 'from', 'K', '##EG', ',', 'T', '##sing', '##hua', 'University', ‘.’, ‘[SEP]’]]

# 构造真正的tensor作为input。return_tensors=‘pt‘指生成pytorch类型的tensor。
batch_input = ["I am Xiao Liu.", "I come from KEG, THU."] 
batch = tokenizer(batch_input, padding=True, truncation=True, return_tensors="pt")
# batch = {'input_ids’: …, 'token_type_ids’: …, 'attention_mask’: …}

outputs = model(**batch)

outputs.last_hidden_state  # torch.Size([2, 12, 768])，两个句子所有token的向量表示
outputs.pooler_output  # torch.Size([2, 768])，两个[CLS] token的向量表示（即句子的裸表示，方法二）
```
